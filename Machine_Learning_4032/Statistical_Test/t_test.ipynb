```python
# For numerical operations
import numpy as np
# For plotting the graph
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the number of samples
ss = 100000
```

```python
# Generate data: Squared values of normal distributions
X1 = np.random.normal(size=ss)**2
X2 = np.random.normal(size=ss)**2 + np.random.normal(size=ss)**2
X3 = np.random.normal(size=ss)**2 + np.random.normal(size=ss)**2 + np.random.normal(size=ss)**2
X4 = np.random.normal(size=ss)**2 + np.random.normal(size=ss)**2 + np.random.normal(size=ss)**2 + np.random.normal(size=ss)**2
```

```python
X30 = 0
for i in range(30):
    X30 += np.random.normal(size=ss)**2
```

```python
# Generate standard normal values
Z = np.random.standard_normal(size=ss)
```

```python
# Compute t-distributions
t1 = Z / np.sqrt(X1 / 1)
t2 = Z / np.sqrt(X2 / 2)
t3 = Z / np.sqrt(X3 / 3)
t4 = Z / np.sqrt(X4 / 4)
t30 = Z / np.sqrt(X30/30)
```

```python
# Plot the distributions
plt.figure(figsize=(10, 6))
sns.kdeplot(t1, label='t-distribution (df=1)', linewidth=2)
sns.kdeplot(t2, label='t-distribution (df=2)', linewidth=2)
sns.kdeplot(t3, label='t-distribution (df=3)', linewidth=2)
sns.kdeplot(t4, label='t-distribution (df=4)', linewidth=2)
sns.kdeplot(t30, label='t-distribution (df=30)', linewidth=2)
sns.kdeplot(Z, label='Standard Normal', linestyle='dashed', linewidth=2)

plt.legend()
plt.title("T-Distributions with Different Degrees of Freedom")
plt.xlabel("Value")
plt.ylabel("Density")
plt.xlim(-5, 5)  # Limiting x-axis
plt.show()
```

```python
# For using t-distribution functions
import scipy.stats as stats  
```

```python
# Define degrees of freedom (df)
df = 4
```

```python
# Generate x values in a specified range for plotting
x = np.linspace(-5, 5, 100)  # Create 1000 values between -5 and 5
```

```python
# Compute the probability density function (PDF) for t-distribution
# Using the probability density function
y_t = stats.t.pdf(x, df)  
```

```python
# Compute the probability density function (PDF) for normal distribution (for comparison)
 # Standard normal distribution (mean=0, std=1)
y_norm = stats.norm.pdf(x) 
```

```python
# Plot the t-distribution
 # Normal distribution in red dashed line
plt.plot(x, y_t, label=f'T-Distribution (df={df})', color='b')
plt.plot(x, y_norm, label='Normal Distribution', color='r', linestyle='dashed') 
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title("Student's t-Distribution vs Normal Distribution")
plt.legend()
plt.grid()
plt.show()
```

```python
# Compute important statistics of the t-distribution
mean, var, skew, kurt = stats.t.stats(df, moments='mvsk')

# Print computed values
print(f"Mean: {mean}")  # Mean of the distribution
print(f"Variance: {var}")  # Variance of the distribution
print(f"Skewness: {skew}")  # Skewness of the distribution
print(f"Kurtosis: {kurt}")  # Kurtosis of the distribution
```
